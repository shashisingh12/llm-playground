```python
import torch
import time

# Set up devices
cpu_device = torch.device("cpu")
if torch.backends.mps.is_available():
    mps_device = torch.device("mps")
    print("MPS is available")
else:
    mps_device = None
    print("MPS not available")

# Create large random tensors
size = (1000, 1000)
a_cpu = torch.randn(size, device=cpu_device)
b_cpu = torch.randn(size, device=cpu_device)

if mps_device:
    a_mps = a_cpu.to(mps_device)
    b_mps = b_cpu.to(mps_device)

# CPU timing
start = time.time()
result_cpu = torch.matmul(a_cpu, b_cpu)
torch.cuda.synchronize() if torch.cuda.is_available() else None  # For CUDA, not needed for CPU
cpu_time = time.time() - start
print(f"CPU matmul time: {cpu_time:.4f} seconds")

# MPS timing
if mps_device:
    # Warm up
    _ = torch.matmul(a_mps, b_mps)
    torch.mps.synchronize()
    start = time.time()
    result_mps = torch.matmul(a_mps, b_mps)
    torch.mps.synchronize()
    mps_time = time.time() - start
    print(f"MPS matmul time: {mps_time:.4f} seconds")
else:
    print("MPS device not available for timing.")
```
